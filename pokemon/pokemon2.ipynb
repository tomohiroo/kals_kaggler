```python
pokemon['Type 2'] = pokemon['Type 2'].fillna('None')
pokemon['Name'] = pokemon['Name'].fillna('Primeape')
pokemon.iloc[[62],:]
```

```python
battle2 = battle.copy()
hp_dic = dict(zip(pokemon['#'], pokemon['HP']))
attack_dic = dict(zip(pokemon['#'], pokemon['Attack']))
defense_dic = dict(zip(pokemon['#'], pokemon['Defense']))
spattack_dic = dict(zip(pokemon['#'], pokemon['Sp. Atk']))
spdefense_dic = dict(zip(pokemon['#'], pokemon['Sp. Def']))
speed_dic = dict(zip(pokemon['#'], pokemon['Speed']))
legend_dic = dict(zip(pokemon['#'],pokemon['Legendary']))
battle2["First_pokemon_hp"] = battle2["First_pokemon"].replace(hp_dic)
battle2["First_pokemon_attack"] = battle2["First_pokemon"].replace(attack_dic)
battle2["First_pokemon_defense"] = battle2["First_pokemon"].replace(defense_dic)
battle2["First_pokemon_spattack"] = battle2["First_pokemon"].replace(spattack_dic)
battle2["First_pokemon_spdefense"] = battle2["First_pokemon"].replace(spdefense_dic)
battle2["First_pokemon_speed"] = battle2["First_pokemon"].replace(speed_dic)
battle2["First_pokemon_legend"] = battle2["First_pokemon"].replace(legend_dic)
battle2["Second_pokemon_hp"] = battle2["Second_pokemon"].replace(hp_dic)
battle2["Second_pokemon_attack"] = battle2["Second_pokemon"].replace(attack_dic)
battle2["Second_pokemon_defense"] = battle2["Second_pokemon"].replace(defense_dic)
battle2["Second_pokemon_spattack"] = battle2["Second_pokemon"].replace(spattack_dic)
battle2["Second_pokemon_spdefense"] = battle2["Second_pokemon"].replace(spdefense_dic)
battle2["Second_pokemon_speed"] = battle2["Second_pokemon"].replace(speed_dic)
battle2["Second_pokemon_legend"] = battle2["Second_pokemon"].replace(legend_dic)
```
