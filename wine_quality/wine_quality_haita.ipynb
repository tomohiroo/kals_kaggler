```python
def wine_quality():
    import numpy as np
    import numpy.random as random
    import scipy as sp
    import pandas as pd
    from pandas import Series, DataFrame
    
    from sklearn import linear_model
    clf = linear_model.LinearRegression()

    # 可視化モジュール
    import matplotlib.pyplot as plt
    import matplotlib as mpl
    import seaborn as sns
    %matplotlib inline

    # 小数第３まで表示
    %precision 3
    
    wine_quality = pd.read_csv("/Users/hatsuisa/Downloads/train.csv")
    
    wine_quality.rename(columns={"fixed acidity" : "fixed_acidity"}, inplace=True)
    wine_quality.rename(columns={"volatile acidity" : "volatile_acidity"}, inplace=True)
    wine_quality.rename(columns={"citric acid" : "citric_acid"}, inplace=True)
    wine_quality.rename(columns={"residual sugar" : "residual_sugar"}, inplace=True)
    wine_quality.rename(columns={"free sulfur dioxide" : "free_sulfur_dioxide"}, inplace=True)
    wine_quality.rename(columns={"total sulfur dioxide" : "total_sulfur_dioxide"}, inplace=True)
    
    print(wine_quality.describe())
    print(wine_quality.info())
    
    print("\n相関係数")

    print("fixed_acidity:", sp.stats.pearsonr(wine_quality.fixed_acidity, wine_quality.quality))
    print("volatile_acidity:", sp.stats.pearsonr(wine_quality.volatile_acidity, wine_quality.quality))
    print("citric_acid:", sp.stats.pearsonr(wine_quality.citric_acid, wine_quality.quality))
    print("residual_sugar:", sp.stats.pearsonr(wine_quality.residual_sugar, wine_quality.quality))
    print("chlorides:", sp.stats.pearsonr(wine_quality.chlorides, wine_quality.quality))
    print("free_sulfur_dioxide:", sp.stats.pearsonr(wine_quality.free_sulfur_dioxide, wine_quality.quality))
    print("total_sulfur_dioxide:", sp.stats.pearsonr(wine_quality.total_sulfur_dioxide, wine_quality.quality))
    print("density:", sp.stats.pearsonr(wine_quality.density, wine_quality.quality))
    print("pH:", sp.stats.pearsonr(wine_quality.pH, wine_quality.quality))
    print("sulphates:", sp.stats.pearsonr(wine_quality.sulphates, wine_quality.quality))
    print("alcohol:", sp.stats.pearsonr(wine_quality.alcohol, wine_quality.quality))
    
    X = wine_quality.loc[:,['alcohol']].as_matrix()
    
    Y = wine_quality['quality'].as_matrix()
    
    clf.fit(X, Y)
    print("回帰係数:", clf.coef_)
    print("切片:", clf.intercept_)
    
    sns.pairplot(wine_quality[["volatile_acidity", "alcohol", "quality"]])
    plt.grid(True)
wine_quality()
```
