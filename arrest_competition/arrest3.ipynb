```python
duration_mapping = {'1-15 min':1, '16-30 min':2, '30+ min':4}
df_train["stop_duration"] = df_train['stop_duration'].map(duration_mapping)
```

```python
df_train["stop_date"] = pd.to_datetime(df_train["stop_date"],format = '%Y-%m-%d %H:%M:%S')
date_year = df_train["stop_date"].dt.year
date_month = df_train["stop_date"].dt.month
```

```python
def to_season(month_num):
    season = "winter"
    if 3<= month_num <= 5:
        season = "spring"
    elif 6<= month_num <= 8:
        season = "summer"
    elif 9<= month_num <= 11:
        season = "autumn"
    return season
```

```python
date_season = pd.Categorical(date_month.apply(to_season),categories = ["spring","summer","autumn","winter"])
date_season
```

```python
week = df_train["stop_date"].dt.dayofweek
df_train["stop_time"] = pd.to_datetime(df_train["stop_time"],format="%H:%M")
time = df_train["stop_time"].dt.hour
```

```python
def to_time(timer):
    t = "early"
    if 6 <= timer <= 11:
        t = "morning"
    elif 12<= timer <= 17:
        t = "noon"
    elif 18 <= timer <= 23:
        t = "evening"
    return t
```

```python
when = pd.Categorical(time.apply(to_time),categories=["early","morning","noon","evening"])
```

```python
a = pd.get_dummies(df_train["driver_race_raw"])
c = pd.get_dummies(df_train["search_conducted"])
d = pd.get_dummies(df_train["search_type"])
e = pd.get_dummies(df_train["contraband_found"])
f = pd.get_dummies(df_train["driver_gender"])
g = pd.get_dummies(df_train["county_fips"])
x = pd.get_dummies(when)
y = pd.get_dummies(date_season)
z = pd.get_dummies(date_year)
l = pd.get_dummies(week)
k = pd.get_dummies(df_train["violation"])
```

```python
c.columns = ["conduc","not_con"]
d.columns = ["consent","inventory","other1"]
e.columns = ["found","not_fou"]
```

```python
df_train["age_log"] = df_train["driver_age_raw"].apply(lambda x: np.log(x/10+1))
df_train = df_train.drop("driver_age_raw",axis=1)
```

```python
df_train = pd.concat([df_train,a,c,d,e,f,g,x,y,z,l,k],axis = 1)
```

```python
df_train = df_train.drop({"fine_grained_location","search_type","search_type_raw","driver_age",},axis=1)
```

```python
df_train = df_train.drop({"stop_date","stop_time"},axis=1)
df_train = df_train.drop({"driver_race_raw","search_conducted","contraband_found","driver_gender","county_fips"},axis=1)
df_train = df_train.drop("location_raw",axis=1)
df_train = df_train.drop("violation",axis=1)
df_train = df_train.drop("state",axis=1)
df_train = df_train.drop("officer_id",axis=1)
df_train = df_train.drop({"driver_race","violation_raw","violation_raw","police_department","county_name"},axis=1)
```
